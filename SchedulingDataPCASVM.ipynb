```
#Luciano Andrade
#!pip install rpy2

import rpy2
import seaborn as sns; sns.set()
import numpy as np

from rpy2.robjects import pandas2ri
pandas2ri.activate()
%reload_ext rpy2.ipython

import pandas as pd
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix


features, target = datasets.load_wine(return_X_y=True)

X_reduced = PCA(n_components=4).fit_transform(features)

X = np.concatenate((X_reduced, np.array([target]).T), axis=1)

#---------------------------------------------------------------------------

%R -i X

%R X <- as.data.frame(X);

%R names(X)<-c("V1","V2","V3","V4","class");

%R library(caret)

#%R install.packages("e1071", repos="http://mirror.its.sfu.ca/mirror/CRAN/")

%R library("e1071")

%R index <- createDataPartition(X$class, p=0.80, list=FALSE);

%R testset <- as.data.frame(X)[-index,];

%R trainset <- as.data.frame(X)[index,];

%R svm_model <- svm(class ~ ., data=trainset, kernel="radial", cost=1, gamma=0.5);

%R pred <- predict(svm_model, newdata = testset);

%R testset <- testset$class

%R -o pred,testset

#---------------------------------------------------------------------------

errorRate = round(np.mean(np.round(pred,0) != testset),3);

print(errorRate)

confusion_matrix(testset, np.round(pred,0))

```
