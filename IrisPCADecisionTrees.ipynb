```
#Luciano Andrade
#!pip install rpy2

import rpy2
import seaborn as sns; sns.set()
import numpy as np

from rpy2.robjects import pandas2ri
pandas2ri.activate()
%reload_ext rpy2.ipython

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.metrics import confusion_matrix

iris = datasets.load_iris()

X_reduced = PCA(n_components=3).fit_transform(iris.data)

X = np.concatenate((X_reduced, np.array([iris.target]).T), axis=1)

#---------------------------------------------------------------------------

%R -i X

%R X <- as.data.frame(X);

%R names(X)<-c("V1","V2","V3","species");

%R library(caret)

%R index <- createDataPartition(X$species, p=0.80, list=FALSE);

%R testset <- as.data.frame(X)[-index,];

%R trainset <- as.data.frame(X)[index,];

%R iris.tree = train(species ~ ., data=trainset, method="rpart", trControl = trainControl(method = "cv"));

%R irisPred = predict(iris.tree, newdata = testset);

%R table(irisPred, testset$species)

%R testsetEspecies <- testset$species

%R -o testsetEspecies

%R -o irisPred

#---------------------------------------------------------------------------

errorRate = round(np.mean(np.round(irisPred,0) != testsetEspecies),3);

print(errorRate)

confusion_matrix(testsetEspecies, np.round(irisPred,0))

```
